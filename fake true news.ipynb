```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset1=pd.read_csv('Fake.csv')
dataset0=pd.read_csv('True.csv')
```

```python
dataset1.insert(4, "Real_or_Fake", 0)
dataset0.insert(4, "Real_or_Fake", 1)
```

```python
dataset1
```

```python
dataset0
```

```python
dataset2=dataset0.append(dataset1)
```

```python
dataset2
```

```python
dataset2['total_text_length'] = dataset2['text'].apply(len)+dataset2['title'].apply(len)

dataset2.head()
```

```python
dataset2['total_text'] = dataset2['title']+dataset2['text']

dataset2.head()
```

```python
len(dataset2['total_text'])
```

```python
dataset2
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus=[]
for i in range (0,20000):
    review=re.sub('[^a-zA-Z]',' ',str(dataset2['total_text'][i]))
    review=review.lower()
    review=review.split()
    ps=PorterStemmer()
    all_stop_word=stopwords.words('english')
    
    review=[ps.stem(word) for word in review if not word in set(all_stop_word)]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(corpus).toarray()
y=dataset2.iloc[0:20000,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
pd.DataFrame(list(zip(y_test, y_pred)), columns=['Actual', 'Predicted'])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
